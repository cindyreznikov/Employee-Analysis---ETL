```python
# Create the employees dataframe
SQL_Query = pd.read_sql_query(
'''select
emp_no,
birth_date,
first_name,
last_name,
gender,
hire_date
from employees''',connection)
```

```python
employees_df = pd.DataFrame(SQL_Query, columns=['emp_no','birth_date', 'first_name', 'last_name', 'gender', 'hire_date'])
employees_df.head()
```

```python
# Create the salaries dataframe
SQL_Query = pd.read_sql_query(
'''select
emp_no,
salary,
from_date,
to_date
from salaries''',connection)
```

```python
salaries_df = pd.DataFrame(SQL_Query, columns=['emp_no','salary', 'from_date', 'to_date'])
salaries_df.head()
```

```python
# Create the titles dataframe
SQL_Query = pd.read_sql_query(
'''select
emp_no,
title,
from_date,
to_date
from titles''',connection)
```

```python
titles_df = pd.DataFrame(SQL_Query, columns=['emp_no','title', 'from_date', 'to_date'])
titles_df.head()
```

```python
min_salary = salaries_df['salary'].min()
min_salary
```

```python
max_salary = salaries_df['salary'].max()
max_salary
```

```python
###  merge the 2 dataframes (salaries and titles)

merged_df = pd.merge(salaries_df, titles_df, on="emp_no")
merged_df
```

```python
# The stmts below sorts the dataframe by salary 
merged_df.sort_values(["salary", "emp_no"], ascending=False, inplace=True)
```

```python
###  Set up bins to determine most common salary Ranges
# Create a new dataframe

bins = [0, 39999, 49999, 59999, 69999, 79999, 89999, 99999, 200000]

group_names = ["<40K", "40k-49999","50K-59999", "60K-69999", "70K-79999", "80K-89999", "90K-99999", "100K+"]

merged_df["Salary Ranges"] = pd.cut(merged_df["salary"], bins, labels=group_names,include_lowest=True)
merged_df
```

```python
# Create a histogram showing common salary ranges for employees

plt.hist(merged_df['Salary Ranges'])
plt.xlabel('Salary Ranges')
plt.ylabel('# Employees')
plt.xticks(rotation='45')
plt.show()
```

```python
#  Create a Groupby Object based on Employee Salary
avg_salary_df = merged_df.groupby("title")["salary"].mean()
avg_salary_df
```

```python
# Use DataFrame.plot() in order to create a bar chart of the data
avg_salary_df.plot(kind="bar", figsize=(20,10))

# Set a title for the chart
plt.title("Average Salary")
plt.xlabel("Job Title")
plt.ylabel("Salary")

plt.show()
plt.tight_layout()
                    
```
